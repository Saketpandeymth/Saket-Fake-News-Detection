```python
dataframe_merge = pd.concat([dataframe_fake, dataframe_true], axis =0 )
dataframe_merge.head(10)
```

```python
dataframe = dataframe_merge.drop(["title", "subject","date"], axis = 1)
```

```python
dataframe.isnull().sum()
```

```python
dataframedataframe = dataframe.sample(frac = 1)
dataframe.head()
```

```python
dataframe.reset_index(inplace = True)
dataframe.drop(["index"], axis = 1, inplace = True)
dataframe.head()
```

```python
def wordopt(t):
    t = t.lower()
    t = re.sub('\[.*?\]', '', t)
    t = re.sub("\\W"," ",t)
    t = re.sub('https?://\S+|www\.\S+', '', t)
    t = re.sub('<.*?>+', '', t)
    t = re.sub('[%s]' % re.escape(string.punctuation), '', t)
    t = re.sub('\n', '', t)
    t = re.sub('\w*\d\w*', '', t)
    return t
```

```python
dataframe["text"] = dataframe["text"].apply(wordopt)
```

```python
x = dataframe["text"]
y = dataframe["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
import numpy as np

# Check for NaN values in y_train
nan_indices = np.isnan(y_train)

# Remove rows with NaN values from both features and target variable
xv_train = xv_train[~nan_indices]
y_train = y_train[~nan_indices]
```

```python
from sklearn.linear_model import LogisticRegression


LR = LogisticRegression()
```

```python
pred_lr=LR.predict(xv_test)
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier


DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train
```

```python
pred_dt = DT.predict(xv_test)
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier


GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(xv_train, y_train)
```

```python
pred_gbc = GBC.predict(xv_test)
GBC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier


RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)
```

```python
pred_rfc = RFC.predict(xv_test)
RFC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```
